```python
import numpy as np

# Student names
students = np.array(["Alice", "Bob", "Charlie", "David", "Eve"])

# Scores in Math, Physics, Chemistry (rows: students, columns: subjects)
scores = np.array([
    [85, 90, 78],   # Alice
    [70, 82, 75],   # Bob
    [92, 88, 94],   # Charlie
    [65, 73, 80],   # David
    [78, 85, 72]    # Eve
])
```

```python
mean = np.mean(scores,axis=0)
print(mean)
```

```python
sum = np.sum(scores,axis=1)
max_index = np.argmax(sum)
max = students[max_index]
print(max)
```

```python
corelation = np.corrcoef(scores,rowvar=False)
print(corelation)
```

```python
max_score = np.max(scores)
min_score = np.min(scores)
normalize = (scores-min_score)/(max_score-min_score)
print(normalize)
```

```python
min_vals = np.min(scores, axis=0)  # [65, 73, 72] (min of each subject)
max_vals = np.max(scores, axis=0)  # [92, 90, 94] (max of each subject)
normalized_scores = (scores - min_vals) / (max_vals - min_vals)
print(normalized_scores)
```

```python
# Extract individual subject columns
math = scores[:, 0]
physics = scores[:, 1]
chemistry = scores[:, 2]

# Manually compute Math vs Physics correlation
cov_math_physics = np.cov(math, physics, ddof=1)[0, 1]  # Covariance
std_math, std_physics = np.std(math, ddof=1), np.std(physics, ddof=1)
r_manual = cov_math_physics / (std_math * std_physics)
print("Manual Math vs Physics correlation:", r_manual)  # Should match 0.658
```

```python
import numpy as np # type: ignore

# Columns: [Area (sq ft), Bedrooms, Bathrooms, Age (years), Price ($)]
houses = np.array([
    [2100, 3, 2, 20, 500000],  # House 1
    [1400, 2, 1, 30, 350000],  # House 2
    [1700, 2, 2, 15, 425000],  # House 3
    [3000, 4, 3, 10, 700000],  # House 4
    [2000, 3, 2.5, 5, 600000], # House 5
    [1800, 3, 2, 12, 480000],  # House 6
    [2500, 4, 3, 8, 650000],   # House 7
    [1900, 3, 2, 20, 520000],  # House 8
    [2200, 3, 2.5, 18, 550000],# House 9
    [1500, 2, 1, 25, 400000]   # House 10
])
```

Calculate the average price of houses with >2 bedrooms.



```python
bedrooms = houses[:,1]
prices = houses[:,-1]
b_g_2 = np.mean(prices[bedrooms>2])
print(f"{b_g_2:.1f}")
```

Compute the correlation between:





i. Area and Price


```python
Area = houses[:,0]
cor_A_P = np.corrcoef(Area,prices)[0,1]
print(f"{cor_A_P}")
```

ii. Bedrooms and Price


```python
cor_B_P = np.corrcoef(bedrooms,prices)[0,1]
print(f"{cor_B_P}")
```

Strongest Corelation

```python
if cor_A_P > cor_B_P:
    print(f"Strongest Corelation is Area")
else:
    print(f"Strongest Corelation is Bedroom")
```

```python
# Extract columns
area = houses[:, 0]
bedrooms = houses[:, 1]
price = houses[:, -1]

# Compute correlations
corr_area_price = np.corrcoef(area, price)[0, 1]
corr_bedrooms_price = np.corrcoef(bedrooms, price)[0, 1]

strongest_correlation = "Area" if abs(corr_area_price) > abs(corr_bedrooms_price) else "Bedrooms"
print("Correlation (Area vs Price):", corr_area_price)
print("Correlation (Bedrooms vs Price):", corr_bedrooms_price)
print("Strongest correlation:", strongest_correlation)
```
