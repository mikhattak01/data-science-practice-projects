```python
IQR = dt_n['Lot Frontage'].quantile(0.75) - dt_n['Lot Frontage'].quantile(0.25)
lower_bound = dt_n['Lot Frontage'].quantile(0.25) - 1.5 * IQR
upper_bound = dt_n['Lot Frontage'].quantile(0.75) + 1.5 * IQR   
print('lower_bound:', lower_bound, 'upper_bound:', upper_bound)
```

```python
n_dt = dt_n[(dt_n['Lot Frontage'] >= lower_bound) & (dt_n['Lot Frontage'] <= upper_bound)]
n_dt.shape
```

Removing Outlier using Z score

```python
sns.displot(n1_dt['Lot Frontage'], kde=True)
plt.show()
```

```python
dt_n['Lot Frontage'].describe()
```

```python
min_range = 69.22459-3*23.365665 #mean - 3*std
max_range = 69.22459+3*23.365665 #mean + 3*std
print('min_range:', min_range, 'max_range:', max_range)
```

negative range means no outlier in that direction

```python
n_dt = dt_n[(dt_n['Lot Frontage'] <= max_range)]
n_dt.shape
```

using Z score

```python
z_score = (dt_n['Lot Frontage'] - 69.22459) / 23.365665 #z_score = (x - mean) / std
n1_dt = dt_n[(z_score >= -3) & (z_score <= 3)] #z_score >= -3 means that the value is within 3 standard deviations below the mean, and z_score <= 3 means that the value is within 3 standard deviations above the mean. So, this condition keeps only those values that are within 3 standard deviations from the mean, effectively removing outliers.
n1_dt.shape
```
