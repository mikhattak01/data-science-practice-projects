```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from collections import Counter
%matplotlib inline
```

```python
budget_x = [1, 2, 3, 4, 5]
revenue_y1 = [1, 4, 9, 16, 25]
revenue_y2 = [2, 3, 5, 7, 11]
plt.plot(budget_x, revenue_y1, 'b.:', label='Year 1')  # 'b.:' means blue color, dot marker, dotted line
plt.plot(budget_x, revenue_y2, color="k", linewidth=2, marker="o", label='Year 2')
plt.title('Movie Budget vs Revenue')
plt.xlabel('Budget (in millions)')
plt.ylabel('Revenue (in millions)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
budget_x = [10, 20, 30, 40, 50]
revenue_y1 = [1, 4, 9, 16, 25]
revenue_y2 = [2, 3, 5, 7, 11]

x_indexes = np.arange(len(budget_x))
width = 0.25

plt.bar(x_indexes + width, revenue_y2, width = width)
plt.bar(x_indexes, revenue_y1, width = width)
plt.title('Movie Budget vs Revenue')
plt.xlabel('Budget (in millions)')
plt.ylabel('Revenue (in millions)')
plt.xticks(ticks=x_indexes, labels=budget_x)
plt.show()
```

```python
data = pd.read_csv('data.csv')
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']

language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';'))

languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()

plt.barh(languages, popularity)

plt.title("Most Popular Languages")
# plt.ylabel("Programming Languages")
plt.xlabel("Number of People Who Use")

plt.tight_layout()

plt.show()
```
